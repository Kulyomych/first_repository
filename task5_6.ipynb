```python
import random
import time

def math_game():
    penalty = 3
    while True:
        a = random.randint(1, 100)
        b = random.randint(1, 100)
        correct_sum = a + b

        user_input = input(f"Новый пример: {a} + {b}? ")

        try:
            user_answer = int(user_input)
        except ValueError:
            print("Пожалуйста, вводите только числа!")
            continue

        if user_answer == correct_sum:
            print("Поздравляю! Ваш ответ правильный")
            break
        else:
            print(f"Неверно! Штрафная пауза: {penalty} секунд")
            time.sleep(penalty)
            penalty += 1
math_game()
```
